__KIVY__

__A. PENGERTIAN KIVY__

Kivy merupakan library Python open source yang digunakan untuk mengembangkan aplikasi mobile dengan user inetrface alami (Natural User Interface). Kivy bisa berjalan pada platform Android, iOS, Linux, OS X dan Windows. Kivy berlisensi MIT.

__B. CARA INSTALASI KIVY__

Sebelum menginstal framework kivy pada library python usahan terlebih dahulu sudah menginstal pip. Dan ini beberapa tahapan yang harus dilakukan untuk mengintal kivy melalui CMD :
1. Ketik pada CMD python –m pip install –upgrade pip wheel setuptools.
2. Kemudian selanjutnya ketik python –m pip install docutils pygments pypiwin32 Kivy.deps.sdl2 kivy.deps.glew pada proses ini nanti akan mendownload sebuah file berukuran kurang lebih 120 mb. Setelah download sudah selesai kemudian ketikkan pada kolom berikutnya Python –m pip install kivy.deps.gstreame.
3. Selanjutnya yaitu proses instalasi kivy dengan cara ketikkan python –m pip install kivy kemudian tunggu sampai proses tersebut selesai dan muncul tulisan successfully installed kivy.

__D. PENJELASAN PROGRAM__

Pada pemrograman PIANO terdapat beberapa class diantaranya adalah :
1. Class Hitam

    class ini digunakan untuk menampilkan button / tombol piano yang berwarna hitam
- def __init__(self)
    
    digunakan untuk mengatur tata letak / koordinat dari button piano hitam
- def value_Cs(self) sampai def value_Ds1(self)
    
    digunakan untuk memasukan suara kedalam button ketika di click

2. Class Putih

    class ini digunakan untuk menampilkan button / tombol piano yang berwarna hitam
- def __init__(self)

    digunakan untuk mengatur tata letak / koordinat dari button piano putih

- def value_C(self) sampai def value_F1(self)

    digunakan untuk memasukan suara kedalam button ketika di click
