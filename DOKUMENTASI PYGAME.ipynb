__PAYGAME__

__A. PENGERTIAN PYGAME__

Pygame adalah serangkaian modul Python lintas-platform yang dirancang untuk menulis video game. Ini termasuk grafik komputer dan pustaka suara yang dirancang untuk digunakan dengan bahasa pemrograman Python.

__B. CARA ISNTALASI PYGAME__

Langkah 1 :  cek versi python.
Langkah 2 : Download file instalasi pygame nya. link: https://pygame.org/downloads
Langkah 3 : instalasi 
            1. Klik 2 kali pada file nya, Kemudian lanjutkan ya, klik “Run anyway”.
            2. Selanjutnya klik "next"
            3. Selanjutnya ganti pilihan instalasinya menjadi “specified hard disk”
               Lalu pada bagian folder yang ada di pojok kiri bawah, kita isi dengan lokasi instalasi file python kita.
            4. Setelah itu klik next dan tunggu hingga proses instalasinya selesai.
